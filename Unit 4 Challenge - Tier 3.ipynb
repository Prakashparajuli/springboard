#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

#url= "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
#properties = pd._ _ _(url_LondonHousePrices, sheet_name='Average price', index_col= None)




url= "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties=pd.read_excel(url,sheet_name='Average price', index_col=None)
properties.head()
```

```python
properties.set_index('Unnamed: 0',inplace=True)
properties.index
```

```python
properties.drop(properties.index[0],inplace=True) #This row is a duplicate information for name of borough, Hence I droped it. 
properties.head()
```

```python
properties.index.name='Date'
properties.columns.name='Borough'
properties.head()
```

```python
properties.columns=properties.columns.str.title()
properties.columns
```

```python
properties=properties.apply(pd.to_numeric)
properties.info()
```

```python
properties.describe().T
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
properties=properties.reset_index()
properties.head()
```

```python
properties['Date']
```

```python
properties['year']=properties['Date'].apply(lambda date:date.year)
```

```python
properties.head()
```

```python
columns=properties.columns
columns
```

```python
no_boroughs=['City Of London', 'Date', 'East Midlands', 'East Of England', 'England','Inner London','London','North East','North West','Outer London','South East','South West','West Midlands','Yorks & The Humber']
```

```python
len(columns)-len(no_boroughs)
```

```python
properties=properties.drop(no_boroughs,axis=1)
properties.columns
```

```python
avgPrice_year=properties.groupby('year').mean()# average price by year 
avgPrice_year
```

```python
avgPrice_year.loc[2020,:].sort_values(ascending=False).head(10)
```

```python
avgPrice_year.loc[2018,:].sort_values(ascending=False).head(10)
```

```python
avgPrice_year.reset_index().plot(x='year',y=['Kensington & Chelsea','Barking & Dagenham','Barnet', 'Westminster'], kind='line', figsize=(12,8))
```

```python
price_ratio_1998_18=(avgPrice_year.loc[2018]/avgPrice_year.loc[1998]).sort_values(ascending=False)
price_ratio_1998_18=price_ratio_1998_18.to_frame().reset_index().rename(columns={'Borough':'Borough',0:'price_ratio_1998_2018'})
price_ratio_1998_18.head()
```

```python
avgPrice_year.loc[[1998,2018],'Hackney']
```

```python
ax=price_ratio_1998_18.plot(x='Borough',kind='bar', figsize=(10,6))
ax.set_ylabel('ratio')
```

```python
top5=['Hackney', 'Waltham Forest', 'Southwark', 'Lewisham','Westminster']
avgPrice_year.reset_index().plot(x='year',y=top5, kind='line', figsize=(12,8))
```

```python
price_ratio_1998_18.head()
```

```python
price_ratio_1995_20=(avgPrice_year.loc[2020]/avgPrice_year.loc[1995]).sort_values(ascending=False)
price_ratio_1995_20=price_ratio_1995_20.to_frame().reset_index().rename(columns={'Borough':'Borough',0:'price_ratio_1995_2020'})
price_ratio_1995_20.head()
```

```python
price_ratio_2010_20=(avgPrice_year.loc[2020]/avgPrice_year.loc[2010]).sort_values(ascending=False)
price_ratio_2010_20=price_ratio_2010_20.to_frame().reset_index().rename(columns={'Borough':'Borough',0:'price_ratio_2010_2020'})
price_ratio_2010_20.head()
```

```python
price_ratio_2019_20=(avgPrice_year.loc[2020]/avgPrice_year.loc[2019]).sort_values(ascending=False)
price_ratio_2019_20=price_ratio_2019_20.to_frame().reset_index().rename(columns={'Borough':'Borough',0:'price_ratio_2019_2020'})
price_ratio_2019_20.head()
```

## Top 10 Borough with high price increase 

```python
top_10=pd.concat([price_ratio_1998_18['Borough'],price_ratio_1995_20['Borough'],price_ratio_2010_20['Borough'],price_ratio_2019_20['Borough']],axis=1)
top_10.columns=['Borough_1998_18','Borough_1995_20','Borough_2010_20','Borough_2019_20']
top_10.head(10)
```

### Conclusion
Based on 1998 and 2018, **Hackney** shows a highest increase in average price. However, looking over the trend of recent increase in price other boroughs like **Westminister, Waltham Forest and Barking & Dagenham** also shows higher increase in addition to Hackney.

Interstingly, the average increase in price of Hackney which is in the list of top two shows low increase in 2020 compared to 2019.

Despite having highest increase in average price of Hackney borough, **Kensington & Chelsea** borough is most expensive borough in both 2018 and 2020.
